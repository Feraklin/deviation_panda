## 3. FLOOR VS CEILING
Let's compare MEAN value of deviation for floor and ceiling. 

Hypothesis 0: There are no significant differences between the data frames for the floor and the ceiling points.

Hypothesis 1: There are significant differences between the data frames for the floor and the ceiling points.

```python
Drawing_plots("boxplot",[data.floor_mean, data.ceiling_mean], labels= ["floor mean", "ceiling mean"]).drawing_plots()
```

We can see a lot of outstanding values. Let's try create histogram plot for lg(degrees) to improve visualisation.

```python
Drawing_plots("boxplot",[np.log10(data.floor_mean), np.log10(data.ceiling_mean)], labels= ["lg of floor mean", "lg of ceiling mean"]).drawing_plots()
```

The graph shows that there are no considerable differences between this data frames. For additional analysis the Mann-Whitney U-test was  carried out (10,000 comparisons, samples of 300 elements). And as a result there are no differences only in 30% of repetitions, which means that it is IMPOSSIBLE to say that the data frames are not different.

## 4. 4 corners VS 6 corners VS 8 corners VS 10 corners

Let's compare MEAN value of deviation for data frames with diffferent amount of corners. 

Hypothesis 0: There are no significant differences between the data frames with diffferent amount of corners.

Hypothesis 1: There are significant differences between the data frames with diffferent amount of corners.

```python
Drawing_plots("boxplot",[data.loc[data.rb_corners == 4]["mean"], data.loc[data.rb_corners == 6]["mean"],data.loc[data.rb_corners == 8]["mean"], data.loc[data.rb_corners == 10]["mean"]], labels=["4 corners", "6 corners", "8 corners","10 corners"]).drawing_plots()
```

Let's try create histogram plot for lg(degrees) to improve visualisation.

```python
Drawing_plots("boxplot",[np.log10(data.loc[data.rb_corners == 4]["mean"]), np.log10(data.loc[data.rb_corners == 6]["mean"]),np.log10(data.loc[data.rb_corners == 8]["mean"]), np.log10(data.loc[data.rb_corners == 10]["mean"])], labels=["4 corners", "6 corners", "8 corners","10 corners"]).drawing_plots()
```

There are differences in the data which depend on the number of corners. In additional, the Kruskal-Wallis test was performed. It demonstrated that the distribution of data across all groups is different.
